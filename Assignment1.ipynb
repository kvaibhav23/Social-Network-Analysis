## Question 2

```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
G= nx.Graph()
nodes = ['A', 'B', 'C', 'D', 'E']
G.add_nodes_from(nodes)
edges = [
    ('A', 'B', 5),
    ('A', 'C', 1),
    ('B', 'C', 3),
    ('B', 'D', 4),
    ('C', 'D', 2),
    ('C', 'E', 6),
    ('D', 'E', 7)
]
G.add_weighted_edges_from(edges)
```

```python
degree_centrality = nx.degree_centrality(G)
weighted_degree_centrality = {n: sum(weight for _, _, weight in G.edges(n, data='weight')) for n in G.nodes()}

print("Degree Centrality:")
for node, centrality in degree_centrality.items():
    print(f"{node}: {centrality:.2f}")

print("\nWeighted Degree Centrality:")
for node, centrality in weighted_degree_centrality.items():
    print(f"{node}: {centrality:.2f}")
```

```python
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 6))
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='skyblue')
edges = G.edges(data=True)
nx.draw_networkx_edges(G, pos, edgelist=edges, width=[d['weight'] for _, _, d in edges], edge_color='gray')
nx.draw_networkx_labels(G, pos)
plt.title("Social Network Graph with Weighted Edges")
plt.show()
```

## Question 3

```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
DirecG = nx.DiGraph()
nodes = ['A', 'B', 'C', 'D', 'E']
DirecG.add_nodes_from(nodes)
edges = [
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'C'),
    ('C', 'A'),
    ('C', 'D'),
    ('D', 'E'),
    ('E', 'B'),
    ('B', 'D')
]
DirecG.add_edges_from(edges)
in_degrees = DirecG.in_degree()
out_degrees = DirecG.out_degree()
print("In-Degree:")
for node, degree in in_degrees:
    print(f"{node}: {degree}")
```

```python
print("\nOut-Degree:")
for node, degree in out_degrees:
    print(f"{node}: {degree}")
```

```python
pos = nx.spring_layout(DirecG)
plt.figure(figsize=(10, 6))
nx.draw_networkx_nodes(DirecG, pos, node_size=700, node_color='skyblue')
nx.draw_networkx_edges(DirecG, pos, edgelist=DirecG.edges(), arrowstyle='-|>', arrowsize=20, edge_color='gray')
nx.draw_networkx_labels(DirecG, pos)
plt.title("Directed Social Network Graph")
plt.show()
```

## Question 4

```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
nodes_component_1 = ['A', 'B', 'C', 'D']
edges_component_1 = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D'), ('A', 'D')]
nodes_component_2 = ['E', 'F', 'G', 'H']
edges_component_2 = [('E', 'F'), ('E', 'G'), ('F', 'G'), ('G', 'H'), ('E', 'H')]
G.add_nodes_from(nodes_component_1)
G.add_nodes_from(nodes_component_2)
G.add_edges_from(edges_component_1)
G.add_edges_from(edges_component_2)
```

```python
connected_components = list(nx.connected_components(G))
component_sizes = [len(component) for component in connected_components]
print("Connected Components:")
for i, component in enumerate(connected_components):
    print(f"Component {i + 1}: {component} (Size: {component_sizes[i]})")
```

```python
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 6))
color_map = []
for node in G:
    if node in nodes_component_1:
        color_map.append('skyblue')
    else:
        color_map.append('lightgreen')
nx.draw_networkx_nodes(G, pos, node_size=700, node_color=color_map)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='gray')
nx.draw_networkx_labels(G, pos)

plt.title("Graph with Multiple Disconnected Components")
plt.show()
```
